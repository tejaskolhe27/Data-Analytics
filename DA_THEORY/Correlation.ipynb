```python
from sklearn.datasets import fetch_california_housing
```

return_X_y=True: This parameter specifies that you want the dataset to be split into features (in this case, housing) and target values (in this case, target) and returned as separate variables. When return_X_y is set to True, housing will contain the feature data, and target will contain the target values.

The feature variables and target values are already decided for this dataset. Please refer to this: https://scikit-learn.org/stable/datasets/real_world.html#california-housing-dataset

```python
housing,target = fetch_california_housing(as_frame=True,return_X_y=True)
housing.head()
```

```python
#correlation of all the columns with each other
housing.corr()
```

```python
# Correlation of two specific columns
corr = housing.corr()
corr['MedInc']['AveRooms']
```

```python
cmap= sns.diverging_palette(30,220,as_cmap=True)
sns.heatmap(corr,vmin=-1.0,vmax=1.0,square=True,cmap=cmap)
plt.show()
```
