```python
import numpy as np
import pandas as pd
from scipy.stats import *  #ttest,probplot
import matplotlib.pyplot as plt
import pylab
import seaborn as sns
```

```python
shoes = pd.read_csv("women_shoe_prices.csv")
shoes.shape
```

```python
shoes.sample(5)
```

```python
shoes['midprices'] = (shoes["prices.amountMax"]+shoes["prices.amountMin"])/2
# plot a qqplot to check normality.if the variable is normally distributed most of the point 
# should be along the central diagonal
probplot(shoes["midprices"],dist = 'norm',plot = pylab)
plt.show()
```

```python
#ttest
pink = shoes[shoes.colors =='Pink']
notpink = shoes[shoes.colors !='Pink']

print(ttest_ind(pink.midprices,notpink.midprices,equal_var =False))
#there is significant price difference between pink shoes and not pink
#shoes as the p value is very small
```
