### QUESTION 2
Draw right visualization graphs for the following (2 marks each):

1. Day-wise AQI of India

2. Outliers in AQI of Cyprus

3. Outliers in AQI of United Kingdom

4. Show relation between AQI of India and Nepal

5. Show relation between AQI of US and Canada

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data_date.csv")
df.head()
```

1. Day-wise AQI of India

```python
ind=df.loc[df['Country']=='India']
ind
```

```python
plt.figure(figsize=(20,12))
plt.bar(ind['Date'],ind['AQI Value'])
plt.title("Day-Wise AQI of India")
plt.xlabel("Dates")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.show()
```

2. Outliers in AQI of Cyprus

```python
cyprus=df.loc[df['Country']=='Cyprus']

plt.boxplot(cyprus['AQI Value'])
plt.title("Cyprus BoxPlot")
plt.show()
```

3. Outliers in AQI of United Kingdom

```python
uk=df.loc[df['Country']=='United Kingdom of Great Britain and Northern Ireland']

plt.boxplot(uk['AQI Value'])
plt.title("United Kingdom BoxPlot")
plt.show()
```

4. Show relation between AQI of India and Nepal

```python
import seaborn as sns
nepal=df.loc[df['Country']=='Nepal']
plt.figure(figsize=(20,12))
sns.lineplot(data=nepal,x=nepal.Date,y=nepal['AQI Value'])
sns.lineplot(data=ind,x=ind.Date,y=ind['AQI Value'])
plt.xticks(rotation=45)
plt.show()
```

```python
sns.distplot(nepal['AQI Value'], hist = False, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 2})
sns.distplot(ind['AQI Value'], hist = False, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 2})
plt.show()
```

5. Show relation between AQI of US and Canada

```python
us=df.loc[df['Country']=='United States of America']
can=df.loc[df['Country']=='Canada']

plt.figure(figsize=(20,12))
sns.lineplot(data=us,x=us.Date,y=us['AQI Value'])
sns.lineplot(data=can,x=can.Date,y=can['AQI Value'])
plt.xticks(rotation=45)
plt.show()
```

```python
sns.distplot(us['AQI Value'], hist = False, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 2})
sns.distplot(can['AQI Value'], hist = False, kde = True,
                 kde_kws = {'shade': True, 'linewidth': 2})
plt.show()
```
