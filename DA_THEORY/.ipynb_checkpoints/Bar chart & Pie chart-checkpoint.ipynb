```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ex5.csv")
df
```

```python
df.plot.bar(x='branch number')
plt.show()
```

```python
#to show subplots
df.plot.bar(x='branch number',subplots = True)
plt.show()
```

```python
#to plot only single column
df.plot.bar(x='branch number',y="sale1")
plt.show()
```

```python
df1=pd.read_csv("ex5.csv",index_col='branch number')
df1
```

```python
df1.plot.pie(y='sale1')
plt.show()
```

```python
#to change size to reduce cluster
df1.plot.pie(y='sale1',figsize=(11,6))
plt.show()
```

```python
#show percentage for each category
df1.plot.pie(y='sale1',figsize=(11,6),autopct="%1.1f%%")
plt.show()
```

```python
df1.plot.pie(y='sale2',figsize=(11,6),autopct="%1.1f%%",rotatelabels=45)
plt.show()
```
