## QUESTION 3
Calculate AR,I,MA and ARIMA and SARIMA for Australia AQI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
df = pd.read_csv("data_date.csv")
aus = df.loc[df['Country']=='Australia']
aus
```

```python
aus["Date_New"]=pd.to_datetime(aus['Date'],dayfirst=True)
aus.drop(['Date'],axis = 1,inplace =True)
aus.set_index(aus['Date_New'],inplace = True)
aus
```

```python
aus.drop(['Date_New'],axis = 1,inplace =True)
aus
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import acf,pacf

pacf = pacf(aus['AQI Value'].dropna())
acf = acf(aus['AQI Value'].dropna())
print("Moving Average Values are:")
print(pacf)
print("Auto Regressive Values are:")
print(acf)
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(aus['AQI Value'])

# Extract and print the ADF test results
adf_statistic = result[0]
p_value = result[1]
critical_values = result[4]

print("ADF Statistic:", adf_statistic)
print("p-value:", p_value)
print("Critical Values:", critical_values)
```

```python
plt.plot(aus['AQI Value'])
plt.title('Original Data')
plt.show()
```

```python
df_diff  = aus['AQI Value'].diff(1)
```

```python
plt.plot(df_diff)
plt.title('Original Data')
plt.show()
```

```python
result = adfuller(df_diff.dropna())

# Extract and print the ADF test results
adf_statistic = result[0]
p_value = result[1]
critical_values = result[4]

print("ADF Statistic:", adf_statistic)
print("p-value:", p_value)
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df_diff, order=(1,1,1))
model_fit = model.fit()
```

```python
X = df_diff.values
size = int(len(X) * 0.66)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
	model = ARIMA(history, order=(1,1,1))
	model_fit = model.fit()
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
```

```python
predictions_df = pd.DataFrame(predictions, columns=['Predictions'])
```

```python
df_diff = df_diff.reset_index(drop=True)
predictions_df = predictions_df.reset_index(drop=True)
df_pred = pd.concat([df_diff, predictions_df], axis=1)
```

```python
df_pred
```

```python
plt.figure(figsize=(12,8))
plt.plot(df_pred['AQI Value'], label='Original ')
plt.plot(df_pred['Predictions'], label='Predicted ')
plt.title('ARIMA Prediction')
plt.legend()
plt.show()
```
