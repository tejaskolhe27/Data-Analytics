```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
#Calculate the correlation coefficient
correlation_coefficient = df['reading score'].corr(df['writing score'])
print("Correlation coefficient=",correlation_coefficient)
```

```python
plt.scatter(df['reading score'],df['writing score'])
plt.xlabel('Reading Score')
plt.ylabel("Writing Score")
plt.title(f"Scatter Plot (Correlation: {correlation_coefficient:.2f})")
plt.show()
```

```python
# Identify potential outliers(Values greater than 2*Standard Deviation)
residuals = df['writing score']-df['reading score']
residuals
```

```python
std_deviation=residuals.std()
outliers=df[abs(residuals)>2*std_deviation]
# Highlight potential outliers on the scatter plot
plt.scatter(outliers['reading score'],outliers['writing score'],color='blue',label='Outliers')
plt.legend()
plt.show()
```

### BAD EXAMPLE OF SCATTER PLOT USED FOR CORRELATION

```python
df = pd.read_csv("loan.csv")
correlation_coefficient = df['income'].corr(df['age'])
print("Correlation coefficient=",correlation_coefficient)
plt.scatter(df['income'],df['age'])
plt.xlabel('income')
plt.ylabel("age")
plt.title(f"Scatter Plot (Correlation: {correlation_coefficient:.2f})")
plt.show()
```

```python
residuals = df['age']-df['income']
residuals
std_deviation=residuals.std()
outliers=df[abs(residuals)>2*std_deviation]
# Highlight potential outliers on the scatter plot
plt.scatter(outliers['income'],outliers['age'],color='blue',label='Outliers')
plt.legend()
plt.show()
```
