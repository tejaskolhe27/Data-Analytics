```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
means=[]
for i in range(100):
    #To generate a random array of 5 values,with values between 0&1
    arr=np.random.rand(5)
    #calculate mean of random sample
    s_mean=np.mean(arr)
    #add s_mean to list 
    means.append(s_mean)
sns.histplot(means,kde=True,bins=100,color='darkblue')
plt.show()
```

```python
# Calculate the mean
mean=sum(means)/len(means)
mean
```

```python
# Plot the mean over the distribution to get a sense of central tendancy
plt.axvline(mean,color='k',linestyle='dashed',linewidth=1)
min_ylim,max_ylim=plt.ylim()
plt.text(mean*1.1,max_ylim*0.9,'Mean:{:.2f}'.format(mean))
plt.title('Sampling distribution of sample means of randomly generated samples (no. of examples=100)')
plt.xlabel('Sample mean')
plt.ylabel('frequency')
plt.show()
```
