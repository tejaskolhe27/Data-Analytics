```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
np.random.seed(33)
norm_sample = np.random.randn(10000)
sns.histplot(x=norm_sample,kde = True)
plt.show()
```

```python
print("mean = ",np.mean(norm_sample))
print("meaidan =",np.median(norm_sample))
print("std = ",np.std(norm_sample))
```

```python
#generate a sample of 10000 data points from right skewed distribution
#first parameter if 0 means no skewness , positive means right skewness,negitive 
#means left skewness
#this is means the generated data will have a +ve skew.

skewed_data_R =stats.skewnorm.rvs(10,size = 10000,loc=0)
sns.histplot(x = skewed_data_R,kde =True)
plt.show
```

```python
skewed_data_L =stats.skewnorm.rvs(-10,size = 10000,loc = 5)
sns.histplot(x = skewed_data_L,kde  = True)
plt.show()
```

```python
print("mean = ",np.mean(skewed_data_L))
print("meaidan =",np.median(skewed_data_L))
print("std = ",np.std(skewed_data_L))
```

```python
# Q-Q Plots 
stats.probplot(norm_sample,plot=plt)
plt.show()
```

```python
stats.probplot(skewed_data_R,plot=plt)
plt.show()
```

```python
stats.probplot(skewed_data_Lprint(stats.skew(norm_sample)),plot=plt)
plt.show()
```

```python
# print the skewness of each dataset
print("Normal: ",stats.skew(norm_sample))
print("Right Skewed: ",stats.skew(skewed_data_R))
print("Left Skewed: ",stats.skew(skewed_data_L))
```

```python
# perform a skewness test on each dataset
#returns a numerical value representing the skewness of dataset
print("Normal: ",stats.skewtest(norm_sample))
print("Right Skewed: ",stats.skewtest(skewed_data_R))
print("Left Skewed: ",stats.skewtest(skewed_data_L))
```

Output: Statistic: The statistic value is the test statistic calculated by the skewness test. It quantifies the degree of skewness in the dataset. If negative, it indicates that the dataset is left-skewed or negatively skewed.


The p-value measures the evidence against a null hypothesis. # A small p-value (typically less than 0.05) suggests that there is strong evidence to reject the null hypothesis, indicating that the data is significantly skewed.


A larger p-value (greater than 0.05) suggests that there is not enough evidence to reject the null hypothesis, indicating that the data may not be significantly skewed.
