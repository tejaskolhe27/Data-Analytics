```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170
rev_stdev = 20
iteration = 1000
```

```python
#create a normal distribution using thee
revenue = np.random.normal(rev_m,rev_stdev,iteration)
print(revenue)
```

```python
plt.figure(figsize = (10,6))
plt.title("Revenue Simulation")
plt.plot(revenue)
plt.show()
```

```python
#now cost of goods sold
#generallt it is about 60% of the revenuue with 10% SD
#we already have 1000 revenue records,use them randomly assign COGS in the above range
COGS = (revenue*np.random.normal(0.6,0.1))
plt.figure(figsize = (10,6))
plt.title("Cost of good sold Simulation")
plt.plot(COGS)
plt.show()
```

```python
#Calculate Gross Profit
Gross_Profit = revenue-COGS
print(Gross_Profit)
```

```python
plt.figure(figsize = (10,6))
plt.title("Gross Profit Simulation")
plt.plot(Gross_Profit)
plt.show()
```

```python
# Create a stcked bar chart
number = list(range(1,1001))
plt.figure(figsize = (10,6))
plt.bar(number,revenue,label = "Revenue",color  = 'skyblue')
plt.bar(number,COGS,bottom = revenue,label = "COGS",color  = 'lightcoral')
plt.bar(number,Gross_Profit,bottom = [r+c for r,c in zip(revenue,COGS)],
        label = "Gross Profit",color  = 'lightgreen')
plt.xlabel("Number of times ")
plt.ylabel("Amount($)")
plt.legend()
plt.title("Revenue,COGS,Gross product using monte Carlo")
plt.show()
```
