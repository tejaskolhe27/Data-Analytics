```python
import pandas as pd
```

```python
election09 = pd.read_csv('LS2009Candidate.csv')
election14 = pd.read_csv('LS2014Candidate.csv')
```

```python
election09['Alliance'] = election09['Party Abbreviation'].copy()
election14['Alliance'] = election14['Party Abbreviation'].copy()
election09
```

```python
UPA = ['INC','NCP', 'RJD', 'DMK', 'IUML','JMM','JD(s)','KC(M)','RLD','RSP'
       ,'CMP(J)','KC(J)','PPI','MD']

NDA = ['BJP','SS', 'LJP', 'SAD', 'RLSP',
'AD','PMK','NPP','AINRC','NPF','RPI(A)','BPF','JD(U)','SDF','NDPP','MNF','RIDALOS','KMDK',
       'IJK','PNK','JSP','GJM','MGP','GFP','GVP','AJSU','IPFT','MPP','KPP','JKPC','KC(T)','BDJS',
       'AGP','JSS','PPA','UDP','HSPDP','PSP','JRS','KVC','PNP','SBSP','KC(N)','PDF','MDPF']
Other = ['YSRCP','AAAP', 'IND', 'AIUDF', 'BLSP', 'JKPDP', 'JD(S)', 'INLD', 'CPI', 'AIMIM',
'KEC(M)','SWP', 'NPEP', 'JKN', 'AIFB', 'MUL', 'AUDF', 'BOPF', 'BVA', 'HJCBL',
'JVM','MDMK']
```

```python
election09['Alliance'].replace(UPA,'UPA',inplace=True)
election09['Alliance'].replace(NDA,'NDA',inplace=True)
election09['Alliance'].replace(Other,'Other',inplace=True)
```

```python
election14['Alliance'].replace(UPA,'UPA',inplace=True)
election14['Alliance'].replace(NDA,'NDA',inplace=True)
election14['Alliance'].replace(Other,'Other',inplace=True)
```

```python
winning09 = election09.loc[election09['Position']==1.0]
winning14 = election14.loc[election14['Position']==1.0]

```
