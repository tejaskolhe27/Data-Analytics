What is the average Purchase Price?

```python
print("The average purchase price is: ",walmart['Purchase Price'].mean())
```

b. What were the highest and lowest purchase prices?

```python
print("The highest purchase price is: ",walmart['Purchase Price'].max())
print("The lowest purchase price is: ",walmart['Purchase Price'].min())
```

c. How many people have English 'en' as their Language of choice on the website?

d. What is the email of the person with the following Credit Card Number:

e. Someone made a purchase that came from Lot: "90 WT" , what was the Purchase
Price for this transaction?

```python
lot = walmart[walmart['Lot']== "90 WT"]
print("The purchase price from lot 90 WT is:"
      ,lot.at[513,'Purchase Price'])
```

f. How many people have American Express as their Credit Card Provider and made
a purchase above $95 ?

```python
p = walmart[(walmart['Purchase Price']>95) & (walmart['CC Provider']=="American Express")]
p
```

```python
print("""
The number of people have American Express as their Credit Card Provider and made a purchase above $95 are: """
      , p['CC Provider'].count())
```

g. How many people have a credit card that expires in 2025?

```python
walmart['CC Exp Date']=pd.to_datetime(walmart['CC Exp Date'],format="%m/%y")
walmart
```

```python
expire25=walmart[(walmart['CC Exp Date']>="2025-01-01") & (walmart['CC Exp Date']<"2026-01-01")]
expire25
```

```python
print("The number of people have a credit card that expires in 2025 are: "
      ,expire25['CC Exp Date'].count())
```
