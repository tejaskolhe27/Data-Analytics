```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

a. Create a figure object and axis of the figure canvas at axis1 = [0,0,1,1]. Provide
necessary labels and title to the plot.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

b. Create another plot at axis1 = [0,0,1,1] and axis2 = [0.2,0.5,0.2,0.2]

```python
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
fig
```

c. Create a plot by adding two axis object at [0,0,1,1] & [0.2,0.5,0.4,0.4]

```python
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

```python
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')


ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)

fig
```

Use x, y, z array to create the plots

```python
plt.plot(x,z)
plt.xlabel("X-Values")
plt.ylabel("Z-Values")
plt.show()
```

 * Between x & y arrays by setting the x-limit in range 20-22 and y-limit range in
30-50

```python
plt.plot(x,y)
plt.xlabel("X-Values")
plt.ylabel("Y-Values")
plt.xlim(20,22)
plt.ylim(30,55)
plt.show()
```
